```python
new_df['duration_num']=new_df['duration_num'].apply(lambda x:x*24 if x<=3 else x)
new_df
```

```python
new_df['type_of_film']=new_df["duration_num"].apply(lambda x: 'short' if x <90 else 'long')
new_df
```
